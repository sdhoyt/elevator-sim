# Elevator Class Demo

This notebook contains a driver showing how the Elevator class is used to run a simulation with max columns and rows set to allow the entire simulation data resulting from the go_to_floor method to be seen.

## Example

import the elevator module and pandas (to set the max columns/rows to display)

```python
import elevator
import pandas as pd
```

This sets the max columns and rows to display to allow for easier visual inspection of the resulting simulation dataframe.

```python
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
```

Create the elevator object and specify the current floor (`current_floor`), elevator speed (`sec_per_floor`), and lowest/highest floors of the building (`min_building_floor` and `max_building_floor`). The below inputs are all the default values for the constructor.

```python
elev = elevator.Elevator(current_floor = 1, sec_per_floor = 10, min_building_floor = 1, max_building_floor = 20)
```

Run the `go_to_floor` method with a desired list of floors, specify `live_sim` as `True` to run a live simulation (set as `True` for demo purposes only, True is the default), and specify the speed that the simulation should output the data (also set for demo purposes only, `1` is the default).

```python
out = elev.go_to_floor([3, 1, 2], live_sim=True, sim_speed = 1)
out
```
